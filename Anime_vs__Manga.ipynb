```python
# Assuming df1 is your DataFrame containing anime data
# Read the CSV file into a DataFrame
df= pd.read_csv('MAL-anime.csv.zip', delimiter=',')

# Filter the DataFrame to include only anime with manga adaptations
anime_with_manga = df[df['Manga'] == 'Yes']

# Now 'anime_with_manga' contains only the rows where the 'Manga' column is 'Yes'
# You can then use this DataFrame for further analysis or processing
```

```python
anime_w_manga = pd.merge(df_anime, df_manga, left_on=['Title'], right_on=['Title'], how='inner', suffixes=('_anime', '_manga'))
anime_w_manga
```

```python
dates = anime_w_manga[['Title','Published', 'Aired']]
dates
```
